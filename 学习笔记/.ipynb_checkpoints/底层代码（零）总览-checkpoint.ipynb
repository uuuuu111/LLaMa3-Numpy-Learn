## 各个层级的任务及实现
#### 变量（Variable）
- 任务：在神经网络中，变量通常指代网络中的参数，例如**权重和偏置**。这些变量需要在网络的训练过程中进行更新，以优化模型的性能。
- 实现：通常使用张量（tensor）或类似的数据结构来**存储**这些变量，它们需要**初始化、更新和存储当前值**，同时支持**梯度的计算以进行反向传播**。

#### 函数（Function）
- 任务：在神经网络中，函数主要指代**激活函数和损失函数**。激活函数用于引入**非线性特性**，而损失函数则用于衡量模型输出与实际标签之间的差异。
- 实现：函数需要**定义其计算过程**，包括**前向传播（计算输出）和反向传播（计算梯度）**。这些函数在模型的不同部分被调用，用于实现网络的非线性映射和误差计算。

#### 层（Layer）
- 任务：层是神经网络的**基本组成单元**，它们接受输入并生成输出。每一层通常**包含一些权重和偏置**，需要实现**前向传播来计算输出**，并且支持**反向传播以更新参数**。
- 实现：每一层都有其**特定的功能和计算方法**。例如，**全连接层**将输入数据与权重矩阵相乘并加上偏置，然后应用激活函数；**卷积层**将输入数据与卷积核进行卷积操作，并应用非线性激活函数等。

#### 模型（Model）
- 任务：模型是由**多个层组成的结构化网络**。它接受输入数据，并通过一系列层进行计算以生成输出。模型的主要任务是将**各层连接起来，管理层之间的数据流动**，并支持**整体的训练和预测功能**。
- 实现：模型通常包含**初始化网络结构**的方法、**定义前向传播的方法（调用各层的前向传播）**、支持**反向传播的方法（根据损失函数计算梯度并更新各层的参数）**，以及其他辅助功能如**保存和加载模型**等。
